(i) In the context of the Hitchcock problem (also known as the transportation problem), an arc (i, j) is called inadmissible if its reduced cost is negative. The reduced cost is computed as:

r(i, j) = c(i, j) - u(i) - v(j)

where c(i, j) is the transportation cost from supply node i to demand node j, and u(i) and v(j) are the dual variables associated with node i and node j, respectively.

We know that the primal-dual algorithm aims to find a feasible solution that satisfies complementary slackness conditions. At each iteration, the algorithm adjusts the dual variables to make inadmissible arcs admissible, ensuring that the reduced costs of all basic variables (arcs with positive flow) are non-negative. When the algorithm terminates, it has found an optimal solution to the transportation problem.

Now, let's prove that only an empty arc can become inadmissible after modification of the dual variables.

Suppose we have an admissible basic arc (i, j) with positive flow (x(i, j) > 0) and reduced cost r(i, j) = 0. If we modify the dual variables, we must ensure that we do not violate the complementary slackness conditions for the basic variables. Therefore, if we increase u(i) by some amount α, we must also increase v(j) by the same amount α to maintain r(i, j) = 0:

r(i, j) = c(i, j) - (u(i) + α) - (v(j) + α) = c(i, j) - u(i) - v(j)

For non-basic arcs with x(i, j) = 0, increasing u(i) by α and v(j) by α will increase their reduced costs by α, which does not make them inadmissible. However, if we increase u(i) by α and do not adjust v(j), then the reduced cost of arc (i, j) will become negative, making it inadmissible.

To construct an example in which an arc does in fact become inadmissible, consider the following transportation problem:

Supply: [3, 4]
Demand: [4, 3]
Cost matrix:
[2, 1]
[1, 2]

In the optimal solution, x(1,1)=2, x(1,2)=1, x(2,1)=2, x(2,2)=2, u=[0, 0], v=[0, 1]. Suppose we increase u(1) by α=1 without adjusting v(1). The reduced costs become:

r(1,1)=-1
r(1,2)=0
r(2,1)=1
r(2,2)=1

Now, the arc (1,1) becomes inadmissible.

We can apply the Hungarian algorithm to find the optimal assignment. The algorithm proceeds as follows:

    Subtract the smallest cost in each row from all the elements in that row.
    Subtract the smallest cost in each column from all the elements in that column.
    Cover the matrix with minimum number of lines (either horizontal or vertical) so that all zeros are covered. If the number of lines equals the size of the matrix, we have found the optimal assignment and can stop.
    If the number of lines is less than the size of the matrix, find the smallest uncovered element and subtract it from all uncovered elements. Add it to all elements covered by two lines.
    Repeat steps 3-4 until the optimal assignment is found.

To show that in a bipartite graph the maximum matching size is the same as the minimum vertex cover size, we can use Konig's theorem, which states that in a bipartite graph, the size of the maximum matching is equal to the size of the minimum vertex cover.

Here's a proof of Konig's theorem that provides an explanation:

    Let G = (A, B, E) be a bipartite graph, where A and B are the disjoint sets of vertices and E is the set of edges between A and B.
    Let M be a maximum matching of G.
    Let R be the set of unmatched vertices in A, and let T be the set of vertices reachable from R by alternating paths (paths that start with an unmatched edge and alternate between matched and unmatched edges).

Now, let's construct the vertex cover:

    Define the set X = (A - T) ∪ (B ∩ T). We claim that X is a vertex cover.

Proof that X is a vertex cover:

    Consider any edge e = (u, v) ∈ E, where u ∈ A and v ∈ B. If u ∈ T, then e cannot be part of the maximum matching M, because if it were, there would be an alternating path from R to u. Therefore, v must be in T and thus v ∈ X.

    If u ∉ T, then u ∈ A - T and thus u ∈ X. In either case, one of the vertices incident to e is in X, so X is a vertex cover.

Now, we need to show that |M| = |X|:

    Observe that every vertex in A - T is matched by M, because if any vertex u ∈ A - T were unmatched, it would be in R, which contradicts our definition of A - T.

    Similarly, every vertex in B ∩ T is also matched by M. Suppose there exists a vertex v ∈ B ∩ T that is unmatched. Then, there must be an alternating path P from R to v, which means P ends with an unmatched edge. But that implies we can increase the size of M by augmenting it along P, which contradicts the fact that M is a maximum matching.

    Since every vertex in X is matched by M, it follows that |M| ≥ |X|.

    Since X is a vertex cover, there cannot be more edges in M than there are vertices in X, so |M| ≤ |X|.

    Combining inequalities from steps 9 and 10, we get |M| = |X|.

Thus, we have shown that in a bipartite graph, the maximum matching size is the same as the minimum vertex cover size, as required.

#### NUMERICAL EXAMPLE

Let's consider a simple bipartite graph G with vertex sets A and B as follows:

A: {a1, a2, a3}
B: {b1, b2, b3}

And let the edges in the graph be:

E: {(a1, b1), (a1, b2), (a2, b2), (a2, b3), (a3, b3)}


Now, let's find a maximum matching and a minimum vertex cover for G.

    Maximum matching: M = {(a1, b1), (a2, b2), (a3, b3)}
    In this case, we can see that every vertex is matched, so the size of the maximum matching |M| = 3.

    Minimum vertex cover: X = {a1, a2, a3} or X = {b1, b2, b3}
    We can see that by choosing all vertices from either set A or set B, we cover all edges in the graph. No smaller subset of vertices can cover all edges. Thus, the size of the minimum vertex cover |X| = 3.

In this example, we see that the maximum matching size (|M| = 3) is the same as the minimum vertex cover size (|X| = 3), which is consistent with the result we proved using Konig's theorem.
